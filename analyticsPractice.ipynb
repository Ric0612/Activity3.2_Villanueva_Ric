```python
import pandas as pd
```

#data preprocessing

```python
loans_df = pd.read_csv("https://raw.githubusercontent.com/datasciencedojo/datasets/refs/heads/master/titanic.csv")
```

```python
display(loans_df)
```

```python
loans_df.info()
```

```python
loans_df.isnull().sum()
```

```python
loans_df.drop_duplicates()
```

```python
loans_df_cleaned = loans_df.drop(['PassengerId', 'Name', 'Ticket', 'Cabin'],axis=1)
```

```python
loans_df_cleaned.head()
```

```python
loans_df_cleaned['Age'] = loans_df_cleaned['Age'].fillna(loans_df_cleaned['Age'].mean())
```

```python
loans_df_cleaned.info()
```

```python
loans_df_cleaned['Embarked'] = loans_df_cleaned['Embarked'].fillna(loans_df_cleaned['Embarked'].mode()[0])
```

```python
loans_df_cleaned.info()
```

```python
loans_df_cleaned['Embarked'] = loans_df_cleaned['Embarked'].astype('category')
loans_df_cleaned['Sex'] = loans_df_cleaned['Sex'].astype('category')
```

```python
loans_df_cleaned.info()
```
